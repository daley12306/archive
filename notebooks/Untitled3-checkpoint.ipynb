```python
with open("/home/jovyan/notebooks/skill_alias_bootstrap.json", "r", encoding="utf-8") as f:
    rule_mapping_raw = json.load(f)

# Key đã là composite "raw|type" từ Untitled2 → dùng trực tiếp
rule_mapping = {}
for k, v in rule_mapping_raw.items():
    key = k.strip().lower()          # "giao tiếp|soft"
    rule_mapping[key] = v["canonical"]  # không thêm |type nữa

print(f"Loaded {len(rule_mapping)} rule entries")
print(f"Tech: {sum(1 for k in rule_mapping if k.endswith('|tech'))}")
print(f"Soft: {sum(1 for k in rule_mapping if k.endswith('|soft'))}")
print("\nSample keys:")
for k, v in list(rule_mapping.items())[:5]:
    print(f"  {k!r:35s} → {v!r}")
```

```python
df_silver = spark.table("nessie.silver.jobs")

tech = (
    df_silver.select(F.explode("tech_skills").alias("raw_skill"))
    .filter(F.col("raw_skill").isNotNull() & (F.col("raw_skill") != ""))
    .withColumn("skill_type", F.lit("tech"))
)
soft = (
    df_silver.select(F.explode("soft_skills").alias("raw_skill"))
    .filter(F.col("raw_skill").isNotNull() & (F.col("raw_skill") != ""))
    .withColumn("skill_type", F.lit("soft"))
)

all_skills = (
    tech.union(soft)
    .withColumn("raw_skill_norm", F.trim(F.lower(F.col("raw_skill"))))
    .groupBy("raw_skill_norm", "skill_type")
    .agg(F.count("*").alias("freq"))
    .orderBy(F.col("freq").desc())
)

skills_pd = all_skills.toPandas()
# KHÔNG dedup — giữ cả (networking, tech) và (networking, soft)
skills_pd["composite_key"] = skills_pd["raw_skill_norm"] + "|" + skills_pd["skill_type"]

print(f"Total (raw, type) pairs: {len(skills_pd)}")
print(skills_pd["skill_type"].value_counts())
```

```python
mapped_keys = set(rule_mapping.keys())
skills_pd["in_rule"] = skills_pd["composite_key"].isin(mapped_keys)

unmapped_df = skills_pd[
    ~skills_pd["composite_key"].isin(mapped_keys) &
    (skills_pd["freq"] >= 5)
].copy()

print(f"Đã có mapping: {skills_pd['in_rule'].sum()}")
print(f"Unmapped + freq>=5: {len(unmapped_df)}")
print(unmapped_df["skill_type"].value_counts())
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

SIMILARITY_THRESHOLD = 0.93
cluster_mapping = {}

if len(unmapped_df) > 0:
    model = SentenceTransformer('intfloat/e5-small-v2')

    # Tách canonical theo type
    tech_canonicals = list(set(
        v for k, v in rule_mapping.items() if k.endswith("|tech")
    ))
    soft_canonicals = list(set(
        v for k, v in rule_mapping.items() if k.endswith("|soft")
    ))

    print(f"Tech canonicals: {len(tech_canonicals)}")
    print(f"Soft canonicals: {len(soft_canonicals)}")

    # Embed canonical pools
    print("Embedding tech canonicals...")
    tech_canon_emb = model.encode(tech_canonicals, normalize_embeddings=True, batch_size=128)
    print("Embedding soft canonicals...")
    soft_canon_emb = model.encode(soft_canonicals, normalize_embeddings=True, batch_size=128)

    # Tách unmapped theo type
    unmapped_tech = unmapped_df[unmapped_df["skill_type"] == "tech"]["raw_skill_norm"].tolist()
    unmapped_soft = unmapped_df[unmapped_df["skill_type"] == "soft"]["raw_skill_norm"].tolist()

    def match_batch(skills, skill_emb, canon_list, canon_emb, skill_type):
        results = {}
        BATCH = 512
        for i in range(0, len(skills), BATCH):
            batch = skill_emb[i:i+BATCH]
            sims = batch @ canon_emb.T
            best_idx = sims.argmax(axis=1)
            best_sim = sims.max(axis=1)
            for skill, idx, sim in zip(skills[i:i+BATCH], best_idx, best_sim):
                if sim >= SIMILARITY_THRESHOLD:
                    results[f"{skill}|{skill_type}"] = {
                        "canonical": canon_list[idx],
                        "type": skill_type,
                        "sim": round(float(sim), 3)
                    }
        return results

    if unmapped_tech:
        print(f"Embedding {len(unmapped_tech)} unmapped tech skills...")
        tech_emb = model.encode(unmapped_tech, normalize_embeddings=True,
                                batch_size=128, show_progress_bar=True)
        cluster_mapping.update(match_batch(
            unmapped_tech, tech_emb, tech_canonicals, tech_canon_emb, "tech"
        ))

    if unmapped_soft:
        print(f"Embedding {len(unmapped_soft)} unmapped soft skills...")
        soft_emb = model.encode(unmapped_soft, normalize_embeddings=True,
                                batch_size=128, show_progress_bar=True)
        cluster_mapping.update(match_batch(
            unmapped_soft, soft_emb, soft_canonicals, soft_canon_emb, "soft"
        ))

    print(f"\nMatched: {len(cluster_mapping)}/{len(unmapped_df)}")
    print("\nSample tech matches:")
    tech_samples = {k: v for k, v in cluster_mapping.items() if v["type"] == "tech"}
    for k, v in list(tech_samples.items())[:5]:
        print(f"  {k!r:35s} → {v}")
    print("\nSample soft matches:")
    soft_samples = {k: v for k, v in cluster_mapping.items() if v["type"] == "soft"}
    for k, v in list(soft_samples.items())[:5]:
        print(f"  {k!r:35s} → {v}")
```

```python
all_mapping = {}

for _, row in skills_pd.iterrows():
    composite = row["composite_key"]
    raw = row["raw_skill_norm"]
    orig_type = row["skill_type"]

    if composite in rule_mapping:
        all_mapping[composite] = {
            "canonical": rule_mapping[composite],
            "type": orig_type,
            "source": "rule"
        }
    elif composite in cluster_mapping:
        all_mapping[composite] = {
            "canonical": cluster_mapping[composite]["canonical"],
            "type": orig_type,
            "source": f"cluster:{cluster_mapping[composite]['sim']}"
        }
    else:
        all_mapping[composite] = {
            "canonical": raw,
            "type": orig_type,
            "source": "none"
        }
```

```python
rows = []
for composite, info in all_mapping.items():
    # Tách composite key lấy lại raw_skill_norm thực sự
    raw = composite.rsplit("|", 1)[0]   # "asset integration|tech" → "asset integration"
    rows.append({
        "raw_skill": raw,
        "canonical_skill": info["canonical"],
        "skill_type": info["type"],
        "source": info["source"]
    })

df_result = pd.DataFrame(rows)
print(df_result["source"].apply(
    lambda s: "cluster" if str(s).startswith("cluster") else s
).value_counts())
print(df_result["skill_type"].value_counts())
print("\nSample:")
print(df_result.head(10)[["raw_skill","canonical_skill","skill_type","source"]].to_string())
```

```python
import json

# Filter cluster theo threshold
def get_sim(s):
    try: return float(str(s).split(":")[1])
    except: return 0.0

df_result["sim"] = df_result["source"].apply(get_sim)
df_result["source_final"] = df_result.apply(
    lambda r: r["source"] if (
        r["source"] == "rule" or
        (str(r["source"]).startswith("cluster") and r["sim"] >= 0.93)
    ) else "none",
    axis=1
)
df_result["canonical_final"] = df_result.apply(
    lambda r: r["canonical_skill"] if r["source_final"] != "none"
              else r["raw_skill"],
    axis=1
)

# Smart title
def smart_title(s):
    if not isinstance(s, str): return s
    words = s.split()
    return " ".join(
        w if (w.isupper() or any(c.isdigit() for c in w)) else w.capitalize()
        for w in words
    )
df_result["canonical_final"] = df_result["canonical_final"].apply(smart_title)

# Export CSV
df_result.to_csv("/home/jovyan/notebooks/skill_mapping_final.csv", index=False)

# Export JSON với type
alias_dict = {
    row["raw_skill"]: {
        "canonical": row["canonical_final"],
        "type": row["skill_type"]
    }
    for _, row in df_result.iterrows()
}

with open("/home/jovyan/notebooks/skill_alias_final.json", "w", encoding="utf-8") as f:
    json.dump(alias_dict, f, ensure_ascii=False, indent=2)

print(f"CSV: {len(df_result)} rows")
print(f"JSON: {len(alias_dict)} entries")
print(f"\nSource distribution:")
print(df_result["source_final"].value_counts())
print(f"\nType distribution:")
print(df_result["skill_type"].value_counts())
print("\nSample JSON:")
for k, v in list(alias_dict.items())[:5]:
    print(f"  {k!r:35s} → {v}")
```
